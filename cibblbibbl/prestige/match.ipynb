```python
def get_api_data_tournament_data(_id):
    filename = f'{_id:0>8}.json'
    p = pathlib.Path(cibblbibbl.settings["cibblbibbl-data.path"]) / f'cache/api-tournament/{filename}'
    if p.is_file() and p.stat().st_size:
        with p.open(encoding="utf8") as f:
            o = json.load(f)
    else:
        o = pyfumbbl.tournament.get(_id)
        if p.parent.is_dir():
            with p.open("w", encoding="utf8") as f:
                json.dump(
                    o,
                    f,
                    ensure_ascii = False,
                    indent = "\t",
                    sort_keys = True,
                )
    return o
```

```python
def get_api_data_tournament_schedule_data(_id):
    filename = f'{_id:0>8}.json'
    p = pathlib.Path(cibblbibbl.settings["cibblbibbl-data.path"]) / f'cache/api-tournament-schedule/{filename}'
    if p.is_file() and p.stat().st_size:
        with p.open(encoding="utf8") as f:
            o = json.load(f)
    else:
        o = pyfumbbl.tournament.schedule(_id)
        if p.parent.is_dir():
            with p.open("w", encoding="utf8") as f:
                json.dump(
                    o,
                    f,
                    ensure_ascii = False,
                    indent = "\t",
                    sort_keys = True,
                )
    return o
```

```python
def iter_match_ids(group_id, session=None):
  """Iterates over all the matchId values of the group."""
  tourneys = pyfumbbl.group.get_tournaments_data(group_id, session=session)
  for t in tourneys:
    schedule = get_api_data_tournament_schedule_data(t["id"])
    for d in schedule:
        id_ = d["result"].get("id")
        if id_:
            yield id_
```

```python
def get_xml_match_data(match_id):
    filename = f'{match_id:0>8}.xml'
    p = pathlib.Path(cibblbibbl.settings["cibblbibbl-data.path"]) / f'cache/xml-match/{filename}'
    if p.is_file() and p.stat().st_size:
        with p.open(encoding="utf8") as f:
            o = ElementTree.fromstring(f.read())
    else:
        text = pyfumbbl.match.get_legacy_data(match_id)
        repat = '<tournamentWeight>\d+?</tournamentWeight>'
        text = re.sub(repat, '', text)
        xml_ = xml.dom.minidom.parseString(text)
        # uniformizing
        nodelist1 = xml_.getElementsByTagName('performances')
        for element in nodelist1:
          nodelist2 = element.childNodes
          fsort = lambda x: int(x.attributes['player'].value)
          nodelist2.sort(key=fsort)
        # prettify
        text = xml_.toprettyxml(encoding='utf8').decode('utf8')
        o = ElementTree.fromstring(text)
        if p.parent.is_dir():
            with p.open("w", encoding="utf8") as f:
                f.write(text)
    return o
```

```python
def get_replay_data(replay_id):
    filename = f'{replay_id:0>8}.json'
    p = pathlib.Path(cibblbibbl.settings["cibblbibbl-data.path"]) / f'cache/replay/{filename}'
    if p.is_file() and p.stat().st_size:
        with p.open(encoding="utf8") as f:
            o = json.load(f)
    else:
        o = fumbblreplay.get_replay_data(replay_id)
        if p.parent.is_dir():
            with p.open("w", encoding="utf8") as f:
                json.dump(
                    o,
                    f,
                    ensure_ascii = False,
                    indent = "\t",
                    sort_keys = True,
                )
    return o
```

```python
points = {
    "loss (conceded)": -10,
    "loss": 0,
    "dindividual": 1,
    "win": 3,
    "win (Crushing Victory)": 4,
}
```

```python
season_teams = {}
season_results = {}
for T in season_tournaments:
    schedule = [d for d in pyfumbbl.tournament.schedule(T["id"]) if d["result"]["id"]]
    for pairing in schedule:
        match_id = pairing["result"]["id"]
        match_data = pyfumbbl.match.get(match_id)
        conceded = match_data["conceded"]
        conceded = (None if conceded == "None" else conceded)
        if conceded not in (None, "None"):
            # conceded is like "Team2" but match data has "team2" key
            conceded = match_data[conceded.lower()]["id"]
            #print([T, pairing, conceded])
        else:
            conceded = None
        pteam_ids = tuple(d["id"] for d in pairing["teams"])
        pteams = {d["id"]: copy.copy(d) for d in pairing["teams"]}
        for d in pairing["result"]["teams"]:
            pteams[d["id"]]["score"] = d["score"]
        for i, pteam_id in enumerate(pteam_ids):
            pteam_name = pteams[pteam_id]["name"]
            season_teams[pteam_id] = pteam_name
            oppo_pteam_id = pteam_ids[1-i]
            oppo_pteam_name = pteams[oppo_pteam_id ]["name"]
            if conceded == pteam_id:
                r = [match_id, oppo_pteam_id, "loss (conceded)"]
            else:
                score, oppo_score = pteams[pteam_id]["score"], pteams[oppo_pteam_id]["score"]
                if oppo_score == 0 and 4 <= score:
                    r = [match_id, oppo_pteam_id, "win (Crushing Victory)"]
                elif oppo_score < score:
                    r = [match_id, oppo_pteam_id, "win"]
                elif oppo_score == score:
                    r = [match_id, oppo_pteam_id, "dindividual"]
                else:
                    r = [match_id, oppo_pteam_id, "loss"]
            season_results.setdefault(pteam_id, []).append(r)
```
