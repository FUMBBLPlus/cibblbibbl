```python
rows = []
for Mu in G.matchups:
    for teamId, d0 in Mu.config["player_performance"].items():
        Te = cibblbibbl.team.Team(int(teamId))
        for playerId1, d1 in d0.items():
            if d1["type"] == "D":
                playerName = d1["name"]
                teamdata = pyfumbbl.team.get_legacy_data(Te.Id, flags=pyfumbbl.team.PAST_PLAYERS)
                matches = []
                for p in teamdata["players"] + teamdata["pastplayers"]:
                    if p["name"] == playerName:
                        matches.append(p)
                playerId2 = ", ".join(p["id"] for p in matches)
                row = [
                    len(rows) + 1,
                    str(Mu.tournamentId) + "-" + Mu.configfilebasename,
                    Mu.match.Id,
                    teamId,
                    playerName,
                    playerId1,
                    playerId2,
                ]
                rows.append(row)
table = texttable.Texttable(max_width=0)
table.set_deco(table.HEADER)
table.add_rows([["Nr", "Matchup", "MatchId", "TeamId", "PlayerName", "playerId1", "playerId2"],] + rows)
print(table.draw())
```
